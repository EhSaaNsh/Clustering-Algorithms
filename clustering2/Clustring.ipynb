```python
km_set={'init':'random' , 'n_init':10 , 'max_iter':300 , 'random_state':42}

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features =scaler.fit_transform(df3)

## 
```

```python
from sklearn.cluster import KMeans

l = []

for k in range(1,11):
    km = KMeans(n_clusters=k , **km_set)
    km.fit(scaled_features)
    l.append(km.inertia_)
    
## entekhab behtarin k
```

```python
plt.style.use('fivethirtyeight')
plt.xticks(range(1,11))
plt.xlabel('number of cluster')
plt.ylabel('inertia')
plt.plot(range(1,11),l)
```

```python
from kneed import KneeLocator
k1=KneeLocator(range(1,11) , l ,curve='convex' , direction='decreasing')
k1.elbow

## behtarin k baray algorithm k-means barabar 4 ast
```

```python
km = KMeans(
    n_clusters=4, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=20
)
y_km = km.fit_predict(scaled_features)

## cluster bandi ba k-means
```

```python
print(y_km)
```

```python
plt.figure(figsize=(15,8))
centroids = km.cluster_centers_
plt.scatter(X,y,c=km.labels_.astype(float))
plt.scatter(centroids[:,1],centroids[:,2] , c='red')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.show()

## plot cluster k-means
```

```python
aff_set = { 'damping':0.9 , 'max_iter':1000 , 'convergence_iter':20 , 'affinity':'euclidean' ,'random_state':40 ,'preference':12.4}
```
